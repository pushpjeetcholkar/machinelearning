```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("D:\\Education\\Machine Learning\\Machine Learning Essentials - Master core ML concepts\\train.csv")
```

```python
data.head()
```

```python
X=data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
X.shape
```

```python
mu = X.mean(axis=0)
sig = X.std(axis=0)
X = (X-mu)/sig
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
best_features = SelectKBest()
```

```python
fit = best_features.fit(X,y)
```

```python
fit.scores_
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
```

```python
featureScores = pd.concat([dfcolumns,dfscores],axis = 1)
featureScores.columns = ['Features','Scores']
```

```python
featureScores = featureScores.sort_values(by ='Scores',ascending=False)
```

```python
featureScores
```

```python
top_10 =list(featureScores[:10]['Features'])
```

```python
top_10
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import cross_val_score
model = LogisticRegression()
model.fit(X, y)
scores = cross_val_score (model, X, y, cv = 10) 
scores.mean()
```

```python
X[top_10]
```

```python
scores = cross_val_score (model, X[top_10], y, cv = 10) 
scores.mean()
```
